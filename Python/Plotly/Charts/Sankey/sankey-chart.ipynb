# Sankey

```python
import plotly.graph_objects as go
import plotly.express as px

import pandas as pd
```

```python
df = pd.read_csv('Datasets/Migration_NZ/migration_nz.csv')
df
```

```python
all_nodes = df.Measure.values.tolist() + df.Country.values.tolist()


source_indices = [all_nodes.index(measure) for measure in df.Measure] ## Retrieve source nodes indexes as per all nodes list.
target_indices = [all_nodes.index(country) for country in df.Country] ## Retrieve destination nodes indexes as per all nodes list.

```

```python
fig = go.Figure(data=[go.Sankey(
                        node = dict(
                              pad = 20,
                              thickness = 20,
                              line = dict(color = "black", width = 1.0),
                              label =  all_nodes,
                        ),

                        link = dict(
                              source =  source_indices,
                              target =  target_indices,
                              value =  df.Value,
                        )
                    )
                ])

fig.update_layout(title_text="Population Migration between New Zealand and Other Continents",
                  font_size=10)
fig.show()
```

```python
# Selecione os dados relevantes
data = [
    dict(
        source=df['Country'],
        target=df['Citizenship'],
        value=df['Value'],
    )
    for _, df in df.groupby(['Year', 'Measure'])
]

# Combine as colunas 'source' e 'target' em uma Ãºnica coluna para os rÃ³tulos dos nÃ³s
nodes = pd.concat([df['Country'], df['Citizenship']]).unique()

# Crie o objeto de grÃ¡fico de Sankey
fig = go.Figure(go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=nodes,
    ),
    link=dict(
        source=[nodes.tolist().index(s) for i in range(len(data)) for s in data[i]['source']],
        target=[len(nodes) + nodes.tolist().index(t) for i in range(len(data)) for t in data[i]['target']],
        value=[val for i in range(len(data)) for val in data[i]['value']],
    ),
))

# Atualize a formataÃ§Ã£o do grÃ¡fico
fig.update_layout(
    title_text="Fluxo de migraÃ§Ã£o",
    font_size=10,
)

# Mostre o grÃ¡fico
fig.show()


```
