# Line Plot with go.Scatter


```python
import plotly.graph_objects as go
```

```python
chinese_american_museu
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=chinese_american_museu['Month'], y=chinese_american_museu['Chinese American Museum']))

annotations = []

annotations.append(dict(xref='paper', yref='paper', x=0.0, y=1.15,
                              xanchor='left', yanchor='bottom',
                              text='Quantidade de visitas ao Chinese American Museum',
                              font=dict(family='Arial',
                                        size=25,
                                        color='black'),
                              showarrow=False))

fig.update_layout(annotations=annotations)

fig.show()
```
