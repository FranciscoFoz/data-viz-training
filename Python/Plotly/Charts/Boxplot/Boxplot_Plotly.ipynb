# Boxplot com Plotly

```python
import pandas as pd
import plotly.express as px
```

## Manipulando os dados

```python
salarios_bibliotecas_california = pd.read_excel('https://github.com/FranciscoFoz/Plotly_graficos/raw/main/Boxplot/California-Public-Libraries-Salary-survey-FY20-21.xlsx',
                                                sheet_name=8,
                                                skiprows=1,
                                                skipfooter=7)
salarios_bibliotecas_california.head()
```

```python
salarios_diretores = salarios_bibliotecas_california.iloc[:,[0,1,2]]
salarios_diretores.columns = ['Biblioteca','Salario inicial','Salario final']
salarios_diretores
```

```python
salarios_diretores.replace('Unavailable',None,inplace=True)
```

```python
salarios_diretores.info()
```

```python
salarios_diretores = salarios_diretores.dropna()
```

```python
salarios_diretores.info()
```

```python
salarios_diretores = salarios_diretores.copy()
```

```python
salarios_diretores.iloc[:,[1,2]] = salarios_diretores.iloc[:,[1,2]].astype(float)
salarios_diretores
```

## Plotando o gráfico

```python
fig = px.box(salarios_diretores)
fig.show()
```

```python
fig = px.box(salarios_diretores,
             labels={'value':'Salário mensal em dólares',
                     'variable':'Tipo de salário'},
             title='Distribuição dos salários dos diretores das bibliotecas públicas da Califórnia-EUA',
             )
fig.show()
```

## Gráfico final

```python
fig = px.box(salarios_diretores,
             labels={'value':'Salário mensal em dólares',
                     'variable':'Tipo de salário'},
             title='Distribuição dos salários dos diretores das bibliotecas públicas da Califórnia-EUA',
             width=900,
             height=600
             )
fig.update_layout(title_font_size=22,
                  font_size=15,
                  xaxis_title=None)
fig.show()
```
