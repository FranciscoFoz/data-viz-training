```python
import pandas as pd
```

```python
df = pd.read_csv('/home/franciscofoz/Documents/GitHub/data-viz-training/Datasets/Relatorio de vendas/relatorio_vendas.csv')
df
```

```python
df.info()
```

```python
df['data_pedido'] = pd.to_datetime(df['data_pedido'])
```

```python
df_cliente_vendas = df.groupby(['nome_cliente'])['vendas'].sum().nlargest(10).copy()
df_cliente_vendas = df_cliente_vendas.reset_index()
df_cliente_vendas
```

```python
df_cliente_vendas['Rank'] = df_cliente_vendas.index + 1
```

```python
df_cliente_vendas.set_index('Rank', inplace=True)
```
