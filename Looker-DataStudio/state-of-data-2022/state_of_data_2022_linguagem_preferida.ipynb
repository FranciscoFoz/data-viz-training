# Linguagem preferida dos profissionais de Dados no Brasil

## Importando bibliotecas

```python
import pandas as pd

pd.options.display.max_columns = 999
```

## Importação dos dados

```python
url = 'Datasets/State_0f_data_2022/State_of_data_2022.csv'

state_of_data = pd.read_csv(url)
```

## Linguagem preferida por cargo e UF

```python
cargo_linguagem_preferida = state_of_data.loc[:,[   "('P1_i_1 ', 'uf onde mora')",
                                                    "('P2_f ', 'Cargo Atual')",
                                                    "('P4_f ', 'Entre as linguagens listadas abaixo, qual é a sua preferida?')"]]
cargo_linguagem_preferida.columns = ['uf', 'cargo_atual', 'linguagem_preferida']
cargo_linguagem_preferida = cargo_linguagem_preferida.fillna('Não informado')
cargo_linguagem_preferida['linguagem_preferida'] = cargo_linguagem_preferida['linguagem_preferida'].str.upper()
cargo_linguagem_preferida
```

## Percentual de linguagem preferida

```python
def crostab_linguagem_uf(df):
    percentual_linguagem_preferida_por_uf = (pd.crosstab(df['linguagem_preferida'], 
                                                    df['uf'],
                                                    normalize='columns') * 100)\
                                                    .round(1).reset_index()
    return percentual_linguagem_preferida_por_uf
```

```python
percentual_linguagem_preferida_por_uf = (pd.crosstab(cargo_linguagem_preferida['linguagem_preferida'], 
                                                    cargo_linguagem_preferida['uf'],
                                                    normalize='columns') * 100)\
                                                    .round(1).reset_index()
percentual_linguagem_preferida_por_uf
```

```python
percentual_linguagem_preferida_por_uf.query('linguagem_preferida == "PYTHON" or linguagem_preferida == "R"')
```

```python
# Aplicando "melt" na tabela e excluindo primeira linha
percentual_linguagem_preferida_por_uf_PYTHON = percentual_linguagem_preferida_por_uf\
                                                                                    .query('linguagem_preferida == "PYTHON"')\
                                                                                    .melt(value_name='percentual')[1:]
# Excluindo linha "Não informado"                                                                                    
percentual_linguagem_preferida_por_uf_PYTHON = percentual_linguagem_preferida_por_uf_PYTHON\
                                                                                    .query('uf != "Não informado"')
# Inserindo 'BR-' antes de cada uf                                                                                    
percentual_linguagem_preferida_por_uf_PYTHON.uf = percentual_linguagem_preferida_por_uf_PYTHON.uf.apply(lambda x: 'BR-' + x)

# Inserindo linguagem preferida por estado
percentual_linguagem_preferida_por_uf_PYTHON['linguagem_preferida'] = 'PYTHON'

percentual_linguagem_preferida_por_uf_PYTHON
```

```python

percentual_linguagem_preferida_por_uf_PYTHON.to_csv('Datasets/State_0f_data_2022/PYTHON.CSV',index=False)
```

### Linguagem Preferida: Estatísticos

```python
# Filtrando linguagem preferida por estatísticos
percentual_linguagem_preferida_por_uf_R = crostab_linguagem_uf(cargo_linguagem_preferida.query('cargo_atual == "Estatístico"'))

# Pivotando tabela
percentual_linguagem_preferida_por_uf_R = percentual_linguagem_preferida_por_uf_R.pivot_table(columns='linguagem_preferida').reset_index()

# Inserindo 'BR-' antes de cada uf      
percentual_linguagem_preferida_por_uf_R.uf = percentual_linguagem_preferida_por_uf_R.uf.apply(lambda x: 'BR-' + x)

# Inserindo linguagem preferida por estado
percentual_linguagem_preferida_por_uf_R['linguagem_mais_preferida'] = ['PYTHON' if x > y else 'R' for x, y in zip(percentual_linguagem_preferida_por_uf_R['PYTHON'], percentual_linguagem_preferida_por_uf_R['R'])]

percentual_linguagem_preferida_por_uf_R
```

```python
percentual_linguagem_preferida_por_uf_R.linguagem_mais_preferida.value_counts(normalize=True).round(2)
```

```python
percentual_linguagem_preferida_por_uf_R.to_csv('Datasets/State_0f_data_2022/ESTATISTICOS.CSV',index=False)                                
```
